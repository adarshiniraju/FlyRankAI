## 1. My lane (or freestyle) and why

*Name your lane — or say 'freestyle' and describe your own question. One short paragraph: why this one?*

```python
### Chosen Lane: Content Refresh & Decay Prevention

In organic search (SEO), content value naturally decays over time due to shifts in user search intent, competitive updates, and search engine algorithmic adjustments. Instead of manually auditing thousands of pages or waiting until production traffic completely drops to zero, we need a proactive machine learning framework to spot which valuable pages are entering a systemic downward trend so an editor can intervene early.

This lane focuses on moving past rigid, flat rules (like checking if a page is simply old) and utilizing an ML framework that dynamically balances traffic volume, current rank position, and velocity of decline to flag high-priority decay targets.
```

## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*

```python
* **The Core Question:** Which high-traffic pages are experiencing a true directional decline in organic search performance and require a content refresh to protect client revenue?
* **Unit of Analysis:** A single page (anonymized `page_id`).
* **The Decision:** Determining whether a page's recent traffic drop is a temporary baseline fluctuation or systemic decay requiring a manual update.
* **The Human Action:** An SEO strategist or content editor reviews the flagged page, updates outdated facts, optimizes for missing subtopics, or improves user experience signals.
* **The Cost of a Wrong Recommendation:**
  * **False Positive (Flagging a healthy page):** Wastes expensive editorial time rewording content that did not need fixing, potentially harming its current stable rankings.
  * **False Negative (Missing a truly decaying page):** The page continues its downward slide unnoticed, resulting in permanent loss of traffic, conversions, and business revenue to competitors.
* **Why it's not just "train a model":** Success isn't maximizing an accuracy score on a balanced spreadsheet. It is about optimizing **Precision@K**—ensuring that if we hand an editor a prioritized queue of 20 or 50 pages to fix this week, those top pages genuinely deserve their attention.
```

## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd
import numpy as np

# Load the starter dataset relative to the work/notebooks/ directory
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

# Calculate Number 1: Total Baseline Decay Rate
df["is_declining"] = df["trend_direction"].str.lower().eq("down").astype(int)
decay_rate = df["is_declining"].mean()

# Calculate Number 2: Word Count Myth Verification
median_wc_down = df[df["is_declining"] == 1]["word_count"].median()
median_wc_up = df[df["is_declining"] == 0]["word_count"].median()

# Calculate Number 3: High-Exposure Pages at Risk (Impressions >= 10,000 and declining)
high_traffic_all = df[df["impressions_90d"] >= 10000]
high_traffic_declining = high_traffic_all[high_traffic_all["trend_direction"].str.lower() == "down"]
pct_high_traffic_decaying = (len(high_traffic_declining) / len(high_traffic_all)) * 100

print(f"--- LIVE DATA EVIDENCE ---")
print(f"1. Overall Page Decay Rate: {decay_rate * 100:.1f}% of all pages are trending downward.")
print(f"2. Word Count Distribution: Declining pages median WC ({median_wc_down:.0f}) vs. Growing pages median WC ({median_wc_up:.0f}).")
print(f"3. High-Exposure Risk: {pct_high_traffic_decaying:.1f}% of high-traffic pages (>=10k impressions) are actively losing ground.")
```
